```python
import numpy as np
import pandas as pd
import random

# https://research.unsw.edu.au/projects/unsw-nb15-dataset

train_original_data = pd.read_csv('/content/drive/MyDrive/dataset/pems2.csv')

# Random add label for missing data
for i in range(train_original_data.shape[0]):
    train_original_data.loc[i, 'label3'] = 1
    
```

```python
df = train_original_data

random_indices_train = np.random.choice(df.index, 522*25, replace = False)

for index in random_indices_train[:522*5]:
    df_1 = df.iloc[index]

    df_1['400001'] = 'NaN'
    df_1['label3'] = 0

    # length +=1
    # df.loc[length] = df_1
    df.loc[index] = df_1

df.to_csv('/content/drive/MyDrive/dataset/PEMS_missing_data/PEMS_missing_400001_05.csv')

for index in random_indices_train[522*5:522*10]:
    df_1 = df.iloc[index]
    
    df_1['400001'] = 'NaN'
    df_1['label3'] = 0

    # length +=1
    # df.loc[length] = df_1
    df.loc[index] = df_1

df.to_csv('/content/drive/MyDrive/dataset/PEMS_missing_data/PEMS_missing_400001_10.csv')

for index in random_indices_train[522*10:522*15]:
    df_1 = df.iloc[index]
    
    df_1['400001'] = 'NaN'
    df_1['label3'] = 0

    # length +=1
    # df.loc[length] = df_1
    df.loc[index] = df_1

df.to_csv('/content/drive/MyDrive/dataset/PEMS_missing_data/PEMS_missing_400001_15.csv')

for index in random_indices_train[522*15:522*20]:
    df_1 = df.iloc[index]
    
    df_1['400001'] = 'NaN'
    df_1['label3'] = 0

    # length +=1
    # df.loc[length] = df_1
    df.loc[index] = df_1

df.to_csv('/content/drive/MyDrive/dataset/PEMS_missing_data/PEMS_missing_400001_20.csv')

for index in random_indices_train[522*20:522*25]:
    df_1 = df.iloc[index]
    
    df_1['400001'] = 'NaN'
    df_1['label3'] = 0

    # length +=1
    # df.loc[length] = df_1
    df.loc[index] = df_1

df.to_csv('/content/drive/MyDrive/dataset/PEMS_missing_data/PEMS_missing_400001_25.csv')
```
